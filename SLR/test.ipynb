```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print(df)
```

```python
x = df[['Hours']]
y = df['Scores']

model = LinearRegression()
model.fit(x,y)
m = model.coef_[0]
c = model.intercept_
print(f"Slope of a straight line: {m}x + {c}")
```

```python
y_pred = model.predict(x)
print(y_pred)
```

```python
y_pred = model.predict(x)
plt.scatter(x,y,label="Actual data")
plt.plot(x,y_pred,label="Line of regression")
plt.xlabel("x")
plt.ylabel("y")
plt.title("SLR")
plt.legend()
plt.grid(True)
plt.show()
```
