# Predict final exam scores based on class absences

```python
import pandas as pd
from narwhals import read_csv
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/PC/Desktop/Data Sets/student+performance/student/student-por.csv",usecols=['absences','G3'],sep=';')
print(df)
```

```python
x = df[['absences']]
y = df['G3']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error

print(f"MAE: {mean_absolute_error(y_pred,y_test)}")
```

```python
plt.scatter(x,y,label="Actual data",color="blue")
plt.plot(x_test,y_pred,label="Predicted values",color="red")
plt.grid(True)
plt.legend()
plt.show()
```
