```python
import numpy as np

from sklearn.linear_model import LinearRegression


X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2,4,5,4,5])

model = LinearRegression()
model.fit(X,Y)

m = model.coef_[0]
c = model.intercept_
print(f"slope for line of regression: {m}")
print(f"intercept for line of regression: {c}")
print(f"formula for line of regression:Y = {m}x + {c}")

import matplotlib.pyplot as plt
y_pred = model.predict(X)
plt.scatter(X,Y,label="Actual data")
plt.plot(X,y_pred,label="Line of regression")
plt.xlabel("x")
plt.ylabel("y")
plt.title("SLR")
plt.legend()
plt.grid(True)
plt.show()

```
